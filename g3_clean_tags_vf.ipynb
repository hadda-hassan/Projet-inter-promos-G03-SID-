```python
# NLTK libraries which allows to tokenize the sentence
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk 
nltk.download('stopwords')
nltk.download('punkt')

import pandas as pd
```

## Cleaning the pos_tag

```python
def convert_str_list(chain: str):
    """ Documentation 
    convert str to list '['société' , 'ville]' -> ['société' , 'ville]
    Paramters : 
        chain : text to convert 
    Out :
        chain : the string converted into a list
    """
    chain = chain.replace("[", "").replace("]", "").split(',')
    for i in range(len(chain)):
        chain[i] = chain[i].replace("\'", "")
    return(chain)
```

```python
# loop art_tags to remove stop_words and hashtags

stop = set(stopwords.words('french'))

for i in range(len(df['art_tag'])):
    if df['art_tag'][i] != None:
        if type(df['art_tag'][i]) is str:
            # to convert str to list
            df['art_tag'][i] = convert_str_list(df['art_tag'][i])
        filtred_sentence = []
        for elem in (df['art_tag'][i]):
            # remove the hashtags
            elem = elem.replace('#', '')
            # tokenize elem
            word_tokens = word_tokenize(elem)
            filtred_sentence.append(
                " ".join([w for w in word_tokens if not w.lower() in stop]))
        df['art_tag'][i] = filtred_sentence
```

## Conversion to Json
