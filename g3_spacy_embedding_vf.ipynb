# Data Importation

```python
df = pd.read_json("df_final_clean_without_lem_v0.json")# Import dataset
df = df[['art_id', 'art_content_clean_without_lem']]# Keep columns : art_id and art_content_clean_without_lem 
```

# Pre-trained model in Spacy

```python
# Load the spacy model which has been installed
nlp = spacy.load('fr')


def vector_spacy(art: str) -> list:
    """Documentation

    Parameters:
      art: the article we want to work on

    Out:
      out: sentence embedding of this article

    """
    art = nlp(art)  # Process an article using the model
    return art.vector  # Get the mean vector for the entire article


df["spacy_vector"] = df["art_content_clean_without_lem"].apply(
    vector_spacy)  # Apply the function to every article
df.head()
```

```python
df1 = pd.read_json("Copie de df_deduplicated_v4.json")
df_final = pd.merge(df1, df, left_on="art_id", right_on="art_id", how="left")
df_final = df_final[["art_id", "spacy_vector"]]
df_final.dropna()
df_final.head(5)
```
